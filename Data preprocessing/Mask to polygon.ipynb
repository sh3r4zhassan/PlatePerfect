```python
from rasterio import features
import numpy as np
from PIL import Image
from skimage import draw
import random
from pathlib import Path
from matplotlib import pyplot as plt
import os
import cv2
```

```python
input_dir = '..\\datasets\\UECFOODPIXCOMPLETE\\data\\UECFoodPIXCOMPLETE\\test\\mask'
output_dir = '..\\datasets\\UECFOODPIXCOMPLETE\\data\\UECFoodPIXCOMPLETE\\test\\labels'

for j in os.listdir(input_dir):
    print(j)
    image_path = os.path.join(input_dir, j)
    # load the binary mask and get its contours
    mask = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # load the nonbinary mask to get its labels which are in Red
    class_label = cv2.imread(image_path)
    class_label = class_label[:,:,2]
    
    _, mask = cv2.threshold(mask, 1, 255, cv2.THRESH_BINARY)

    H, W = mask.shape
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # convert the contours to polygons
    polygons = []
    new_countours=[]
    labels=[]
    for cnt in contours:
        if cv2.contourArea(cnt) > 100:
            new_countours.append(cnt)
            polygon = []
            for point in cnt:
                x, y = point[0]
                polygon.append(float(x) / float(W))
                polygon.append(float(y) / float(H))
            
            polygons.append(polygon)
            
    for i in range(len(new_countours)):
        labels.append(class_label[new_countours[i][0][0][1],new_countours[i][0][0][0]])
    print(labels)

    with open('{}.txt'.format(os.path.join(output_dir, j)[:-4]), 'w') as f:

        for i, polygon in enumerate(polygons):          
            for p_, p in enumerate(polygon):
                if p_ == len(polygon) - 1:
                    f.write('{}\n'.format(p))
                elif p_ == 0:
                    f.write('{} {} '.format(labels[i],p))
                else:
                    f.write('{} '.format(p))
                
        f.close()
```
